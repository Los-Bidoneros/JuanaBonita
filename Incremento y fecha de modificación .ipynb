# Incrementar en 1 cada vez que un script se ejecuta 

```python
import pandas as pd
from pandas import ExcelWriter
from pandas import ExcelFile

def conteo_ejecucion(fila,columna,path):
    #Parametros: fila/index, Nombre de la columna, dirección del archivo excel
    df = pd.read_excel(path)
    valor = df.iloc[fila][columna]
    valor = valor+1
    df.at[fila,columna] = valor
    df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
    df.to_excel(path, index=False)
```

```python
# EJEMPLO -------------
conteo_ejecucion(14,"Conteo","C:\\Users\\jhernandez\\Desktop\\Carrpeta\\Scrapers_testeo_modFecha.xlsx")
```

# Insertar fecha de modificacion

```python
import pandas as pd
from pandas import ExcelWriter
from pandas import ExcelFile
from datetime import date
def ultima_mod(fila,columna,path):
    hoy = str(date.today())
    #Parametros: fila/index, Nombre de la columna, dirección del archivo excel
    df = pd.read_excel(path)
    df.at[fila,columna] = hoy
    df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
    df.to_excel(path, index=False)
```

```python
ultima_mod(14,"Ultima modificacion","C:\\Users\\jhernandez\\Desktop\\Carrpeta\\Scrapers_testeo_modFecha.xlsx")
```
